```python
#notebook used to pull VIX data and split into high and low vol periods
import yfinance as yf
import pandas as pd
from datetime import datetime
```

```python
def fetchRange(start, end):
    df = yf.download('^VIX', start=start, end=end, interval='60m')
    return df
```

```python
df_temp = fetchRange(start="2024-03-01", end="2026-02-20")
df_simple = df_temp[['Close']].copy()
df_simple = df_simple.reset_index()   # makes Datetime a column
df_simple.head()
```

```python
df = fetchRange(start="2024-03-01", end="2026-02-20")
print(len(df))
df.head()

mean_price = df['Close'].mean()
upper_threshold = 1.1*mean_price
lower_threshold = 0.9*mean_price

low_vol_values = df[df['Close'] < lower_threshold].dropna()
high_vol_values = df[df['Close'] > upper_threshold].dropna()

print(low_vol_values.mean())
print(high_vol_values.mean())
```
